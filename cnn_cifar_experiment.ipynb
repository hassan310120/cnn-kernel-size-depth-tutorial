```python
# Robust CNN experiment (CIFAR-10):

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load CIFAR-10

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# CIFAR labels
y_train = y_train.squeeze()
y_test  = y_test.squeeze()

# Normalize
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

class_names = [
    "airplane","automobile","bird","cat","deer",
    "dog","frog","horse","ship","truck"
]

# tf.data pipeline (robust + fast)
AUTOTUNE = tf.data.AUTOTUNE
BATCH_SIZE = 128

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_ds = (
    train_ds
    .cache()
    .shuffle(10000)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)

valid_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test))
valid_ds = (
    valid_ds
    .cache()
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)

# CNN model
# -------------------------------
def make_cnn(
    input_shape=(32, 32, 3),
    num_classes=10,
    kernel_size=(3, 3),
    depth=4,
    base_filters=32,
    dropout=0.20,
    use_batchnorm=True,
    use_data_augmentation=True
):
    inputs = keras.Input(shape=input_shape)

    # Data augmentation
    if use_data_augmentation:
        aug = keras.Sequential([
            layers.RandomFlip("horizontal"),
            layers.RandomTranslation(0.10, 0.10),
        ], name="data_augmentation")
        x = aug(inputs)
    else:
        x = inputs

    filters = base_filters

    for i in range(depth):
        x = layers.Conv2D(filters, kernel_size, padding="same")(x)
        if use_batchnorm:
            x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)

        # Downsample every 2 blocks
        if i % 2 == 1:
            x = layers.MaxPooling2D((2, 2))(x)
            x = layers.Dropout(dropout)(x)
            filters = min(filters * 2, 256)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs, name="cnn_cifar")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=5e-4),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

# experimental settings
# -------------------------------
KERNEL_SIZE = (3, 3)
DEPTH = 4

model = make_cnn(kernel_size=KERNEL_SIZE, depth=DEPTH, use_data_augmentation=True)
model.summary()

# Callbacks
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

ckpt_path = f"best_cnn_k{KERNEL_SIZE[0]}_d{DEPTH}.weights.h5"

ckpt = ModelCheckpoint(
    filepath=ckpt_path,
    save_weights_only=True,
    save_best_only=True,
    monitor="val_loss",
    mode="min",
    verbose=1
)

lr_cb = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    min_lr=1e-6,
    verbose=1
)

es = EarlyStopping(
    monitor="val_loss",
    patience=4,
    restore_best_weights=True,
    verbose=1
)

# Train
EPOCHS = 30

history = model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=EPOCHS,
    callbacks=[ckpt, lr_cb, es],
    verbose=1
)

model.load_weights(ckpt_path)

# Training vs validation plots
plt.figure()
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Training vs Validation Accuracy (CIFAR-10)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])
plt.show()

plt.figure()
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Training vs Validation Loss (CIFAR-10)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Train", "Validation"])
plt.show()

# Evaluate on test set
test_loss, test_acc = model.evaluate(valid_ds, verbose=0)
print(f"\nFinal test accuracy: {test_acc:.4f} | test loss: {test_loss:.4f}")

# Experiment on 5 images
# -------------------------------
idx = np.random.choice(len(x_test), size=5, replace=False)
images = x_test[idx]
true_labels = y_test[idx]

pred_probs = model.predict(images, verbose=0)
pred_labels = np.argmax(pred_probs, axis=1)

plt.figure(figsize=(12, 3))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i])     # RGB image
    plt.axis("off")

    true_name = class_names[true_labels[i]]
    pred_name = class_names[pred_labels[i]]
    color = "green" if pred_labels[i] == true_labels[i] else "red"
    plt.title(f"T: {true_name}\nP: {pred_name}", color=color)

plt.suptitle(f"CNN on CIFAR-10 | kernel={KERNEL_SIZE}, depth={DEPTH}")
plt.tight_layout()
plt.show()
```
